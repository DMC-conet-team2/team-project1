```python
import os
import json
import pandas as pd

# 1. JSON 파일들이 들어있는 폴더 경로
folder_path = r'C:\new'  # ← 여기에 실제 경로 입력
output_csv = 'qa_new.csv'  # 저장할 CSV 파일 이름

# 2. 데이터 저장용 리스트
data_rows = []

# 3. 모든 JSON 파일 반복하며 질문과 답변 추출
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)

                # 4. 질문과 답변 추출
                question = data['dataSet']['question']['raw']['text']
                answer = data['dataSet']['answer']['raw']['text']
                file_id = filename.replace('.json', '')

                # 5. 리스트에 추가
                data_rows.append({
                    'id': file_id,
                    'question': question,
                    'answer': answer
                })

        except Exception as e:
            print(f"⚠️ 오류 발생: {filename} → {e}")

# 6. pandas DataFrame으로 변환 후 저장
df = pd.DataFrame(data_rows)
df.to_csv(output_csv, index=False, encoding='utf-8-sig')

print(f"✅ 질문과 답변을 {output_csv} 파일로 저장 완료!")
```
